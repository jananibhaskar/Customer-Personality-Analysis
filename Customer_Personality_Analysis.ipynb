```python
#Dropping Missing Values
df = df.dropna(subset= ['Response','Income'])
df.isnull().sum()
```

```python
from datetime import datetime
# Convert to datetime safely
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], dayfirst=True, errors='coerce')

# Compute customer tenure based on today's date
today = pd.to_datetime(datetime.today().date())
df['Customer_Tenure'] = (today - df['Dt_Customer']).dt.days

```

```python
# Total children
df['Kidhome'] = df['Kidhome'].apply(pd.to_numeric, errors='coerce')
df['Teenhome'] = df['Teenhome'].apply(pd.to_numeric, errors='coerce')
df['Total_Children'] = df['Kidhome'] + df['Teenhome']

# Total spending
spend_cols = ['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']
# Convert all spend columns to numeric (inplace)
df[spend_cols] = df[spend_cols].apply(pd.to_numeric, errors='coerce')

# Now sum them row-wise
df['Total_Spending'] = df[spend_cols].sum(axis=1)

```

Clustering Customers

```python
features = df[['Income', 'Recency', 'Total_Spending', 'NumWebPurchases',
               'NumCatalogPurchases', 'NumStorePurchases', 'NumDealsPurchases',
               'NumWebVisitsMonth', 'Customer_Tenure', 'Total_Children']]

#In order to include marital_status in features we need to convert from categories to numeric for clustering

# One-hot encode marital status
marital_dummies = pd.get_dummies(df['Marital_Status'], prefix='Marital')

# Concatenate the one-hot encoded columns to the features
features=pd.concat([features, marital_dummies], axis=1)
```

Step 5: Clustering Using Machine Learning (K-Means)

1. Scaling Features
2. Finding Optimal Clusters (Elbow & Silhouette)
3. Applying KMeans
4. Interpreting Cluster Sizes

```python
#1. Scale Features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
#2. Elbow Method (to find the best k)
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

wcss = []  # within-cluster sum of squares
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method For Optimal K')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
#3. Silhouette Score (to confirm best k)
from sklearn.metrics import silhouette_score

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(scaled_features)
    score = silhouette_score(scaled_features, labels)
    print(f"Silhouette Score for {k} clusters: {score:.3f}")
```

```python
# 4. Apply K-Means
kmeans = KMeans(n_clusters=10, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_features)
```

```python
#Check how many customers are in each cluster:
df['Cluster'].value_counts()
```

```python
# Assuming 'Cluster' and 'Response' are in your dataframe (Highest response rate)
df.groupby('Cluster')['Response'].mean().sort_values(ascending=False)
```

```python
df['Cluster'] = kmeans.labels_
```

```python
cluster_summary = df.groupby('Cluster').agg({
    'Total_Spending': 'mean',
    'Income': 'mean',
    'NumCatalogPurchases': 'mean',
    'NumStorePurchases': 'mean',
    'NumDealsPurchases': 'mean',
    'NumWebPurchases': 'mean',
    'Recency': 'mean',
    # Add any other columns you think help describe customer behavior
}).reset_index()

print(cluster_summary)
```

```python
#Use PCA for 2D Visualization
from sklearn.decomposition import PCA
import seaborn as sns

pca = PCA(n_components=2)
pca_components = pca.fit_transform(scaled_features)

df['PCA1'] = pca_components[:, 0]
df['PCA2'] = pca_components[:, 1]

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='PCA1', y='PCA2', hue='Cluster', palette='Set2')
plt.title('Customer Segments (PCA Projection)')
plt.show()
```
